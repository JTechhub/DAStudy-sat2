```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('C:/Users/ehdgh_3kgo85z/Desktop/DAstudy/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

## 인터넷서비스, 계약, 지불방법에 따른 이탈율

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

plt.figure(figsize=(8, 5))
churn_counts = df['Churn'].value_counts(normalize=True)
sns.barplot(x=churn_counts.index, y=churn_counts.values)
plt.title('Customer Churn Rate')
plt.xlabel('Churn')
plt.ylabel('Proportion')


services = ['InternetService', 'Contract', 'PaymentMethod']
plt.figure(figsize=(18, 5))

for i, service in enumerate(services, 1):
    plt.subplot(1, 3, i)
    churn_rate = df.groupby(service)['Churn'].value_counts(normalize=True).unstack()
    churn_rate.plot(kind='bar', stacked=True, color=['#4D3425','#E4512B'], ax=plt.gca())
    plt.title(f'Churn Rate by {service}')
    plt.xlabel(service)
    plt.ylabel('Proportion')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## 사용기간, 월별요금과 이탈 여부

```python

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df[df['Churn'] == 'Yes']['tenure'], bins=30, color='red', kde=True)
plt.title('Distribution of Tenure for Churned Customers')
plt.xlabel('Tenure (months)')
plt.ylabel('Number of Customers')

plt.subplot(1, 2, 2)
sns.histplot(df[df['Churn'] == 'No']['tenure'], bins=30, color='green', kde=True)
plt.title('Distribution of Tenure for Retained Customers')
plt.xlabel('Tenure (months)')
plt.ylabel('Number of Customers')

plt.tight_layout()
plt.show()


plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(df[df['Churn'] == 'Yes']['MonthlyCharges'], bins=30, color='red', kde=True)
plt.title('Distribution of MonthlyCharges for Churned Customers')
plt.xlabel('Monthly Charges ($)')
plt.ylabel('Number of Customers')

plt.subplot(1, 2, 2)
sns.histplot(df[df['Churn'] == 'No']['MonthlyCharges'], bins=30, color='green', kde=True)
plt.title('Distribution of MonthlyCharges for Retained Customers')
plt.xlabel('Monthly Charges ($)')
plt.ylabel('Number of Customers')

plt.tight_layout()
plt.show()
```
